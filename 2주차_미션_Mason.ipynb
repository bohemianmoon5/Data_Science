```python
# 미션 데이터

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/medical_201909.csv.zip", low_memory=False)
df.shape
```

```python
# 1번 문제

# 상권업종소분류명에서 약국만 분류
pharm_count = df["상권업종소분류명"] == "약국"

# 시도명으로 빈도수 출력
df.loc[pharm_count, "시도명"].value_counts()
```

```python
# 2번 문제

# 상권업종소분류명에서 동물병원만 분류
sort = df[df["상권업종소분류명"] == "동물병원"]

# 시도명으로 빈도수 출력
count = sort["시도명"].value_counts()
count
```

```python
# 3번 문제

# 시도명에서 서울특별시만 분류
df_seoul = df[df["시도명"] == "서울특별시"]

# 상권업종소분류명에서 피부, 성형이 들어간 것만 분류
sort_df_seoul = df_seoul[df_seoul["상권업종소분류명"].str.contains("피부|성형")]

# 전체 Count
total_count = df_seoul[df_seoul["상권업종중분류명"] == "병원"].groupby("시군구명").size()

# 부분 Count
sort_count = sort_df_seoul.groupby("시군구명").size()

# 반올림 및 정렬
ratio = (sort_count / total_count).round(2)
sort_ratio = ratio.sort_values(ascending=False)
sort_ratio
```

```python
# 4번 문제

import folium

# 상권업종소분류명에서 노인, 치매가 포함된 데이터 분류
sort_df_hospital = df[df["상권업종소분류명"].str.contains("노인|치매")]

# 위도, 경도 평균값
lati = sort_df_hospital["위도"].mean()
long = sort_df_hospital["경도"].mean()

# map 지정
map = folium.Map(location=[lati, long], zoom_start=7)

# 상호명, 도로명주소가 아이콘 정보에 포함되게 map 표시
for idx, row in sort_df_hospital.iterrows():
    name = sort_df_hospital.loc[idx, "상호명"]
    address = sort_df_hospital.loc[idx, "도로명주소"]
    popup = f"{name}-{address}"
    folium.Marker(
        location=[row["위도"], row["경도"]],
        popup=popup,
        icon=folium.Icon(icon="info-sign")
    ).add_to(map)
map
```
